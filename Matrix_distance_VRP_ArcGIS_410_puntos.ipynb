```python
# Function that a given set of points constructs a features.FeatureSet.from_dict
def create_point_dict(df, ref):
    point_dict = { "features":[], "spatialReference": None, "geometryType": None, "fields": []}

    for i, row in df.iterrows():
        point = {}
        X = row["X"]
        Y = row["Y"]
        point["attributes"] = {"Name": str(i + 1), "CurbApproach": 1, "PickupQuantities": 1, "ServiceTime": 1} #1 Minute
        point["geometry"] = {"x": X, "y": Y}
        point_dict["features"].append(point)
    
    point_dict["spatialReference"] = {"wkid": ref, "latestWkid": ref}
    point_dict["geometryType"] = "esriGeometryPoint" 
    point_dict ["fields"] = [{"name" : "Name", "type" : "esriFieldTypeString", "alias" : "Name", "length" : "50"},
                             {"name" : "CurbApproach", "type" : "esriFieldTypeInteger", "alias" : "CurbApproach"},
                             {"name" : "PickupQuantities", "type" : "esriFieldTypeInteger", "alias" : "PickupQuantities"},
                             {"name" : "ServiceTime", "type" : "esriFieldTypeInteger", "alias" : "ServiceTime"}]
    
    return point_dict 
```
