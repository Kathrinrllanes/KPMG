```python
#Fuction for removing nesting and repeated values of a list
def clean_lineString (L):
    
    #Remove nesting
    not_nested_list = [item for sublist in L for item in sublist]

    #Delete consecutive equal values in the list "not_nested_list"
    from itertools import groupby
    clean_list = [x[0] for x in groupby(not_nested_list)]

    return clean_list
```

```python
clean_shape = clean_lineString(shape_route)
```
