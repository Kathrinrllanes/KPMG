```python
# Function that a given set of points constructs a features.FeatureSet.from_dict
def create_point_dict(df, ref):
    point_dict = { "features":[], "spatialReference": None, "geometryType": None, "fields": []}

    for i, row in df.iterrows():
        point = {}
        X = row["X"]
        Y = row["Y"]
        #point["attributes"] = {"Name": str(i + 1), "CurbApproach": 1, "PickupQuantities": 1, "ServiceTime": 1} #1 Minute
        point["attributes"] = {"Name": str(i), "CurbApproach": 1, "PickupQuantities": 1, "ServiceTime": 1} #1 Minute
        point["geometry"] = {"x": X, "y": Y}
        point_dict["features"].append(point)
    
    point_dict["spatialReference"] = {"wkid": ref, "latestWkid": ref}
    point_dict["geometryType"] = "esriGeometryPoint" 
    point_dict ["fields"] = [{"name" : "Name", "type" : "esriFieldTypeString", "alias" : "Name", "length" : "50"},
                             {"name" : "CurbApproach", "type" : "esriFieldTypeInteger", "alias" : "CurbApproach"},
                             {"name" : "PickupQuantities", "type" : "esriFieldTypeInteger", "alias" : "PickupQuantities"},
                             {"name" : "ServiceTime", "type" : "esriFieldTypeInteger", "alias" : "ServiceTime"}]
    
    return point_dict 
```

Como esos shapes son cadenas de lineString anidadas, pues las desanido y elimino los puntos consecutivos iguales con la función clean_lineString

```python
def clean_lineString (L):
    
    #Para concatenar sublistas dentro de una lista L eliminando la anidacion
    not_nested_list = [item for sublist in L for item in sublist]

    #Delete consecutive equal values in the list "not_nested_list"
    from itertools import groupby
    clean_list = [x[0] for x in groupby(not_nested_list)]

    return clean_list
```

# Creating the output dataframes

```python
#Función que dado un arreglo de identificadores de contenedores, devuelve su localización(latitud , longitude)
import utm
def get_location (pos_route):
    route = []
    for i in range(len(pos_route)):
        if pos_route[i] == "TruckBase":
            #Sistema geodésico de referencia: ETRS89 & Proyección cartográfica: UTM Huso 30 N
            #Convert from utm to latitude longitude
            latlon = utm.to_latlon(base_X, base_Y, 30, 'N') #(base de camiones)
        elif pos_route[i] == "Dump":
            latlon = utm.to_latlon(dump_X, dump_Y, 30, 'N') #(vertedero)
        else:
            X = pd_containers.iloc[int(pos_route[i])]['X']
            Y = pd_containers.iloc[int(pos_route[i])]['Y']
            latlon = utm.to_latlon(X, Y, 30, 'N')
        route.append(latlon)
    return route
```

```python
#Función que dado un dataframe que contiene una columna que incluye el identificador de cada contenedor(posicion en el fichero pd_containers)
#Retorna un pandas series con la latitude longitud de cada contenedor

def convert_column_latlon(df):
    #Get only the column of the name of the container
    container_ids = df['Name'].values #the column "Name" contains containers ids

    #Buscar la localizacion de cada contenedor
    container_loc = get_location(container_ids)
    
    #Create a pandas series for the column Localizacion_Contenedor
    #loc_series = pd.Series(latlon_list, name = "Localizacion_Contenedor")
    loc_series = pd.Series(container_loc, name = "Localizacion_Contenedor")
    
    return loc_series
    #return latlon_list
```

```python
#Creating the column "Localizacion_Contenedor"
loc_r1 = convert_column_latlon(orders_r1)
loc_r2 = convert_column_latlon(orders_r2)
loc_r3 = convert_column_latlon(orders_r3)
loc_r4 = convert_column_latlon(orders_r4)
loc_r5 = convert_column_latlon(orders_r5)
```

```python
#Column Damanda
def get_column_demanda (df):
    #Get only the column of the 'PickupQuantities' of the container
    demand = df['PickupQuantities'].values # the column PickupQuantities contains the demand for each container
    demand[0] = '0'
    demand[len(demand)-1] = '0'
    demand_series = pd.Series(demand, name = 'Demanda')
    return demand_series
```
